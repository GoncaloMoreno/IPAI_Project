```python
import os
from astroquery.mast import Observations
from astropy.coordinates import SkyCoord
import pandas as pd
```

can download jwstqueries.csv from https://drive.google.com/file/d/1m_FjXLMdUBJ0_jZaz6F099bea78WuROJ/view?usp=sharing

```python
# Proxima Centauri's ICRS
coord = SkyCoord('14h29m42.94613s', '-62d40m46.16468s', frame='icrs')

jwst_csv = 'jwstqueries.csv'

if os.path.exists(jwst_csv):
    jwst_pd = pd.read_csv(jwst_csv)
else:
    # Query JWST near Proxima Centauri
    obs_table = Observations.query_criteria(
        obs_collection='JWST',  # Hubble Space Telescope
        dataproduct_type=['image', 'spectrum', 'catalog'],  # Data type you want
        s_ra=coord.ra.deg,  # Right Ascension of the region
        s_dec=coord.dec.deg,  # Declination of the region
        radius=0.1  # Search radius in degrees
)
    jwst_pd = obs_table.to_pandas()
    jwst_pd.to_csv('jwstqueries.csv', index=False)
```
